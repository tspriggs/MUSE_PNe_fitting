```python
# moffat model

FWHM = [2.,3.,4.,5.,]
beta = [0.5, 1., 3., 10.]

coordinates = [(n,m) for n in range(30) for m in range(30)]

x_fit = np.array([item[0] for item in coordinates])
y_fit = np.array([item[1] for item in coordinates])

rad = np.sqrt((x_fit - 15.)**2 + (y_fit - 15.)**2)
plt.figure()

for F in FWHM:
    gamma = F / (2. * np.sqrt(2.**(1./beta[1]) - 1.))
    rr_gg = ((np.array(x_fit) - 15.)**2 + (np.array(y_fit) - 15.)**2) / gamma**2
    moffat = 1 * (1 + rr_gg)**(-beta[1])
    
    plt.plot(rad, moffat, label="FWHM = {:}".format(F))
    plt.legend()

plt.figure()
for b in beta:
    gamma = FWHM[1] / (2. * np.sqrt(2.**(1./b) - 1.))
    rr_gg = ((np.array(x_fit) - 15.)**2 + (np.array(y_fit) - 15.)**2) / gamma**2
    moffat = 1 * (1 + rr_gg)**(-b)
    
    plt.plot(rad, moffat, label="beta = {:}".format(b))
    plt.legend()
```
